## Lógica de arreglo

Se implementa un programa que, dado un N ingresado por el usuario, calcule e imprima
el siguiente arreglo:


```python
# Pedir al usuario que ingrese el valor hasta donde se calcula se serie
N = int(input("Ingrese valor de N: "))

def calcular_arreglo(N):
    cifra = 0
    for i in range(1, N + 1):
        cifra = cifra * 10 + i
        resultado = cifra * 8 + i
        print(f"{cifra} * 8 + {i} = {resultado} (N={i})")

calcular_arreglo(N)
```
